##**Exercice1: Etude de la première colonne**
1. **Question 1:**

```r
# let's begin with plotting the time series of X1 variable:
plot(serieX1,main='time series X1')
```

**D'aprés le graphe**
1. La moyenne semble n'est pas constante vu qu'elle varie en fonction du temps
2. La variance varie également (En effet, les fluctuations change clairement en fonction du temps)
3. Clairement, on voit une composante **trend** dans le graphe de time series 
**-->tous ces points renforcent l'hypothèse de la non stationarité de la série temporelle**

```r
# let's try to plot the ACF function to verify the stationarity of the time series:
acf(serieX1)
```

**On remarque qu'il ya beaucoup de composante non nulles qui atteignent des valeurs élévées quoique ça decroit en fonction du temps.**
**Donc, il ya une forte correlation dans le data ce qui renforcent l'hypothèse de la non stationarité de cette série temporelle**

**2.Question 2:**


```r
# on teste la stationarité avec un test de Dicky Fuller: Hypothèse nulle: la série est non stationnaire
adf.test(serieX1)
```

**L'adf test donne un p-value égale à 0.4538>0.05**


**L'hypothèse nulle H0 :"la série n'est pas stationnaire" est retenue**

**Question 3:**

```r
# test de Jung_Box:
Box.test(serieX1, fitdf=0,type ="Ljung-Box" ,lag = 50)
```

**p-value trés petite (inférieur à 0.05)**


**Conclusion : on rejète l'hypothèse nulle( c'est un bruit blanc) et donc il ne s'agit pas d'un bruit blanc, c'est à dire qu'il y a une  correlation dans ce data et le processus est non stationnaire**

**d.S'agit il d'un ARMA(4,3)**

 1. Graphiquement, on peut pas arriver à déterminer l'ordre du modèle vu les graphes acf et pacf obtenues qui ne donnent aucune idée sur p ou q

      **2. En utilisant ARIMA :**


```r
modele_serieX2=arima(serieX2,order=c(4,0,3))
modele_serieX2
```

```r
#let's check the residuals to make sur if it is a white noise or not and then conclude if our model ARMA(4,3) is valid
residus=modele_serieX2$residuals
plot(residus)
```

```r
Box.test(residus,type='Ljung-Box',fitdf=0)
```

p-value = 0.98 > 0.05 donc on retenue l'hypothèse nulle

---> le residu est un bruit blanc

le modèle ARMA(4,3) peut etre un bon candidat pour notre série temporelle


**e. pacf et acf peuvent ils proposer un modèle et un ordre?**

   La serie pourrait donc être modélisée par un processus ARMA(p,q), mais les caractéristiques des fonctions ACf et PACF ne nous donnent pas non plus assez d'informations sur les valeurs de p et q.

```r
auto.arima(serieX2)
```

A l'aide de la fonction autoarima, le modèle le plus approprié à cette série temporelle est **un ARMA(1,2)**

**On va faire une analyse des residus et voir s'ils forment un bruit blanc ou pas**

```r
nouveau_modele=arima(serieX2,order=c(1,0,2))
nouveaux_residus=nouveau_modele$residuals
```

```r
Box.test(nouveaux_residus,fitdf=0,type='Ljung-Box')
```

---> on a un p-value trés grande presque prés de 1 (p-value=0.9891) donc il s'agit d'un bruit blanc et le modèle proposé(ARMA(1,2)) est un bon candidat

####**Question5- Evaluation des paramètres**

**P(|Z|>t)= P(Z>t) + P(-Z>t) = 2 * P(Z>t) = 2*(1 - P(Z<t))**


**P(|Z|>t) = 2*(1 - P(Z<t)) = p_value**

##### **HO : le paramètre est statistiquement nul**
##### **H1 :  le paramèttre n'est pas statistquement nul**


```r
#Modèle ARMA(1,2)
modelARMA12 <- arima(serieX2, order <- c(1, 0, 2))
modelARMA12
```

```r
# installing coeftest :
install.packages('grf')
if(packageVersion("grf") < '0.10.2') {
  warning("This script requires grf 0.10.2 or higher")
}
library(lmtest)
```

```r
#paramètre ar1
t_ar1<- coeftest(modelARMA12)[1] / coeftest(modelARMA12)[5]
pvalue_ar1 <-2*(1-pnorm(abs(t_ar1)))
pvalue_ar1
#Au regard de la p_value très petite, ce paramètre est significatif
```

```r
#paramètre ma1
t_ma1<- coeftest(modelARMA12)[2] / coeftest(modelARMA12)[6]
pvalue_ma1 <-2*(1-pnorm(abs(t_ma1)))
pvalue_ma1
##Au regard de la p_value très petite, ce paramètre est significatif
```

```r
#paramètre ma2
t_ma2<- coeftest(modelARMA12)[3] / coeftest(modelARMA12)[7]
pvalue_ma2 <-2*(1-pnorm(abs(t_ma2)))
pvalue_ma2
#Au regard de la p_value très petite, ce paramètre est significatif
```

**Tous les paramètres du modèle ARMA(1,2) sont significatifs. Le modèle ARMA(1,2) est un bon modèle avec un critère aic = 545.47**


**On a la p-value_ma2 est le plus grand. Supprimons ce paramètre et évaluons un modèle ARMA(1,1)**


```r
#Modèle ARMA(1,1)

#Modèle ARMA(1,1)
modelARMA11 <- arima(serieX2, order <- c(1, 0, 1))
modelARMA11
#On remarque que le crière aic = 551,47. Ce qui est nettement plus grand que le critère aic du modèle ARMA(1,2)
```
